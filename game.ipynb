```python
import cv2

import mediapipe as mp

import random

import os
```

```python
class HandDetector():
    def __init__(self):
        self.mpHands = mp.solutions.hands
        self.hands = self.mpHands.Hands()
        self.mpDraw = mp.solutions.drawing_utils

    def Find_hands(self, img):
        hand_lms = []
        #chuyển ảnh từ BGR thành RGB
        imgRGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        #đưa vào thư viện mediapipe
        results = self.hands.process(imgRGB)
        if results.multi_hand_landmarks:
            for handlm in results.multi_hand_landmarks:
                self.mpDraw.draw_landmarks(img, handlm, self.mpHands.HAND_CONNECTIONS)
                #trích ra các tọa độ của khớp các ngón tay
                firsthand = results.multi_hand_landmarks[0]
                h, w, z = img.shape # chú ý chỗ này
                for i, lm in enumerate(firsthand.landmark):
                    real_x, real_y = int(lm.x * w), int(lm.y*h)
                    hand_lms.append([i, real_x, real_y])
        return img, hand_lms

    def Count_finger(self, hand_lms):
        finger_start_index = [4, 8, 12, 16, 20]
        n_finger = 0
        if len(hand_lms) > 0:
            #kiểm tra ngón cái
            if hand_lms[finger_start_index[0]][1] > hand_lms[finger_start_index[0]-1][1]:
                n_finger += 1
                
            for i in range(1,5):
                if hand_lms[finger_start_index[i]][2] < hand_lms[finger_start_index[i]-2][2]:
                    n_finger += 1
            return n_finger
        else:
            return -1
```

```python
Handdetector = HandDetector()
```

```python
def draw_results(frame, user_draw):
    #sinh ra 1 biến ngẫu nhiên
    random_draw = random.randint(0,2)
    #vẽ hình theo user_draw
    frame = cv2.putText(frame, 'Bạn', (50, 50), cv2.FONT_HERSHEY_COMPLEX, 1, (0,250,0), 2, cv2.LINE_AA)
    s_img = cv2.imread(os.path.join('pic',str(user_draw) + ".png"))
    x_offset = 50
    y_offset = 100
    s_img_resized = cv2.resize(s_img, (240, int(s_img.shape[0] * 240 / s_img.shape[1])))
    frame[y_offset:y_offset+s_img_resized.shape[0], x_offset:x_offset+s_img_resized.shape[1]] = s_img_resized
    
    #vẽ hình theo random_draw
    frame = cv2.putText(frame, 'AI', (400, 50), cv2.FONT_HERSHEY_COMPLEX, 1, (0,0,250), 2, cv2.LINE_AA)
    s_img = cv2.imread(os.path.join('pic',str(random_draw) + ".png"))
    x_offset = 400
    y_offset = 100
    s_img_resized = cv2.resize(s_img, (240, int(s_img.shape[0] * 240 / s_img.shape[1])))
    frame[y_offset:y_offset+s_img_resized.shape[0], x_offset:x_offset+s_img_resized.shape[1]] = s_img_resized
    #kiểm tra và hiển thị kết quả
    if user_draw == random_draw:
        result = 'Hòa'
    elif user_draw == 2 and random_draw == 1:
        result = 'Bạn Thắng'
    elif user_draw == 1 and random_draw == 0:
        result = 'Bạn Thắng'
    elif user_draw == 0 and random_draw == 2:
        result = 'Bạn Thắng'
    else:
        result = 'Bạn Thua'
    frame = cv2.putText(frame, result, (50,500), cv2.FONT_HERSHEY_DUPLEX, 1, (255,0,0), 2, cv2.LINE_AA )
    return frame

cam = cv2.VideoCapture(0)

while True:
    ret, frame = cam.read()
    frame = cv2.flip(frame, 1)
    #đưa hình ảnh vào
    frame, hand_lms = Handdetector.Find_hands(frame)
    n_finger = Handdetector.Count_finger(hand_lms)
    user_draw = -1
    if n_finger == 0:
        user_draw = 0
        print('Cái búa')
        print(n_finger)
    elif n_finger == 2:
        user_draw = 2
        print('Cái Kéo')
        print(n_finger)
    elif n_finger == 5:
        user_draw = 1
        print('Cái bao')
        print(n_finger)
    elif n_finger != -1:
        print('Hãy ra kéo búa hoặc bao')
        print(n_finger)
    else:
        print('Không tìm thấy bàn tay')
        print(n_finger)

    cv2.imshow("game", frame)
    key = cv2.waitKey(1)
    if key == ord('q'):
        print('Trò chơi kết thúc')
        break
    elif key == ord(' '):
        frame1 = draw_results(frame, user_draw)
        cv2.imshow('game',frame1)
        cv2.waitKey()
cam.release()
cv2.destroyAllWindows()
```
